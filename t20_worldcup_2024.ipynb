```python
import pandas as pd
import json
```

```python
with open(r'C:\Data analytics\t20_worldcup_2024\match_details.json') as f:
    data = json.load(f)

df_matches = pd.DataFrame(data)
df_matches
```

```python
df_matches.rename({'scorecard':'match_id'},axis =1,inplace = True)
```

```python
df_matches.head()
```

```python
df_matches.to_csv(r'C:\Data analytics\t20_worldcup_2024\match_details.csv', index = False)
```

```python
match_ids_dict ={}
for index, row in df_matches.iterrows():
   key1 = row['team1']+' Vs '+row['team2']
   key2 = row['team2']+' Vs '+row['team1']

   match_ids_dict[key1] = row['match_id']
   match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
with open(r'C:\Data analytics\t20_worldcup_2024\batting_summary.json') as f:
    data1 = json.load(f)

data1
df_battingsummary = pd.DataFrame(data1)
df_battingsummary
```

```python
df_battingsummary['out/not_out'] = df_battingsummary['dismissal'].apply(lambda s:"not_out" if s == "N/A" else "out")
```

```python
df_battingsummary
```

```python
df_battingsummary.drop(columns = ['dismissal'], inplace = True)
```

```python
df_battingsummary
```

```python
df_battingsummary['batsmanName'] = df_battingsummary['batsmanName'].apply(lambda a:a.replace('†',''))
```

```python
df_battingsummary
```

```python
df_battingsummary['match_id'] = df_battingsummary['match'].map(match_ids_dict)
```

```python
df_battingsummary.head(20)
```

```python
df_battingsummary.to_csv(r'C:\Data analytics\t20_worldcup_2024\batting_summary.csv',index = False)
```
